```python
ranger=range(1,9)
points=[]

for k in ranger:
    model=KMeans(n_clusters=k)
    model.fit(dataset)
    points.append(model.inertia_)

plt.plot(ks,points,'-o')
plt.xlabel('Clusters')
plt.ylabel('Rate')
plt.xticks(ks)
plt.show()
# rate of change from 3-4 is steadily decrementing lets take k=3
```

```python
model=KMeans(n_clusters=3)
prediction=model.fit_predict(dataset.drop('target', axis=1))
centroids=model.cluster_centers_
centroids
```
